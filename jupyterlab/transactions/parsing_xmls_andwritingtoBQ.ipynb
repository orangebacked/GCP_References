```python
from google.cloud import bigquery
import glob
import json
import xmltodict
```

```python
def upload_to_bq(row):
            # Instantiates a client
    bigquery_client = bigquery.Client()
    dataset_ref = bigquery_client.dataset('tries')
    table_ref = dataset_ref.table('try2')
    table = bigquery_client.get_table(table_ref)
    rows_to_insert = [
            row
    ]
    errors = bigquery_client.insert_rows_json(table, rows_to_insert)
    print(errors)
    assert errors == []
```

```python
def recursive_items(dictionary):
    for key, value in list(dictionary.items()):
        new_key = key.replace("#","").replace("@","").replace(':','')
        if new_key != key:
            dictionary[new_key] = dictionary[key]
            del dictionary[key]
        if type(value) is dict:
            recursive_items(value)
        elif type(value) is list:
            for val in value:
                recursive_items(val)
        elif type(value) == str:
            pass
        else:
             recursive_items(value)
    return dictionary

def ordereddict_to_dict(value):
    for k, v in value.items():
        if isinstance(v, dict):
            value[k] = ordereddict_to_dict(v)
    return dict(value)
```

```python
lis_XML = glob.glob("*.XML")
```

```python
for n,file in enumerate(lis_XML):
    print(file)
    with open(file) as fd:
        doc = str(fd.read())
        doc = doc.replace(":","")
        doc = xmltodict.parse(doc)
    aaas = dict(doc)
    namej = '/home/jupyter/newxmls/d/data_{}.json'.format(file.split(".")[0])
    with open(namej, 'w') as outfile:
        json.dump(aaas, outfile)

    x= open(namej, 'r').read().replace("#","").replace("@","")

    with open('/home/jupyter/newxmls/c/{}.json'.format(file.split(".")[0]), 'w') as outfile:
        outfile.write(x)
```
